
=========================================================
Principal components analysis (PCA)
=========================================================

These figures aid in illustrating how a point cloud
can be very flat in one direction--which is where PCA
comes in to choose a direction that is not flat.




```python
print(__doc__)
from sklearn.decomposition import PCA

from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
import os
os.getcwd()
```

```python
# Import data
dataset = pd.read_csv('/Users/xling/Dropbox/MORB K-Rich/PCA/2017 AGU/PCA/marStandardized.csv')
isotopes = dataset.iloc[:,4:9].values
isotopes
```

```python
# PCA
pca = PCA ()
pca.fit(isotopes)
# by default: PCA(copy=True, iterated_power="auto", n_components = all, random_state = None, svd_colver="auto", tol=0.0, whiten=False)
print ("The explained_variance_ratio are:", pca.explained_variance_ratio_)
# get_precision(): compute data precision matrix with the generative model
pca.get_precision()
print ("the", pca.get_precision())
```

```python
score = pca.score_samples(isotopes)
score
```
